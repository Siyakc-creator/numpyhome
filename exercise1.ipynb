```python
import numpy as np
# Columns: [Age, Math Marks, Science Marks],
data = np.array([
       [18, 85, 78],   
       [19, 92, 88],   
        [17, 76, 95],  
        [18, 65, 70],   
        [20, 90, 85]    
    ])
   
```

```python
# get the shape of the matrix
data.shape
```

```python
# find the average age of students
np.mean(data[:,0])
```

```python
# extract math marks of all students 
data[:,1]
```

```python
# find the highest science marks.
np.max(data[:,2])
```

```python
# get details of the students who scored more than 90 in maths
# indexing in boolen



data[data[:,1]>90]
```

```python
# increase maths marks of all students by 5
data[:,1]+=5

data
```

```python
# find how many students are younger than 19
len(data[data[:,0]<19])
```
